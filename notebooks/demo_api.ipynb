# Demo API - Padaria Bom Pão
Este notebook demonstra chamadas à API e indica quais requisitos são validados.

```python
import requests
BASE = 'http://127.0.0.1:5000/api'
# 1) Criar produto válido (valida requisito: preço > 0)
r = requests.post(f"{BASE}/products", json={"sku":"P001","name":"Pão Francês","price":0.8,"quantity":100}, headers={'X-API-KEY':'minha_chave_secreta'})
print('status', r.status_code, 'json', r.json())
```

```python
# 2) Listar produtos (paginação simples)
r = requests.get(f"{BASE}/products")
print(r.json())
```

```python
# 3) Criar cliente (valida CPF e e-mail)
r = requests.post(f"{BASE}/customers", json={"name":"Joao","cpf":"12345678909","email":"joao@mail.com"})
print(r.status_code, r.json())
```

```python
# 4) Criar pedido (valida existência de produto e estoque suficiente)
payload = {"customer_id": 1, "items":[{"product_id":1,"quantity":2}], "seller":"Vendedor1"}
r = requests.post(f"{BASE}/orders", json=payload)
print(r.status_code, r.json())
```
